```python
traj = md.load('10_chains.dcd', top='10_chains.pdb')
```

```python
pairs = [(n, n+1) for n in range(traj.n_atoms - 1)]
dist = md.compute_distances(traj, pairs)
```

```python
data = dist.flatten()
```

```python
scaled = np.int16(data/np.max(np.abs(data)) * 32767)
sound_file = 'test.wav'
wavfile.write(filename=sound_file, rate=44100//10, data=scaled)
```
